```python
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_csv("titanic.csv")
```

```python
df_raw.head()
```

```python
df_raw.shape
```

```python
df = df_raw.copy()
```

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
df.drop(columns=["Cabin"], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```
